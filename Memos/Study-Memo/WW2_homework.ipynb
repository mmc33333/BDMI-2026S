## I: Numpy Practice:

```python
import numpy as np
import numpy.random as npr

a = npr.rand(16)
print(f"原始数组：{a}")

a_matrix = a.reshape(4,4)
print(f"转换为矩阵：{a_matrix}")

a_matrixT = a_matrix.T
print(f"转置后：{a_matrixT}")

I = np.ones((4,4))
ans1 = a_matrixT + I
print(f"加上全一矩阵：{ans1}")
ans2 = a_matrixT - I
print(f"减去全一矩阵：{ans2}")
ans3 = a_matrixT * I
print(f"乘以全一矩阵：{ans3}")
ans4 = a_matrixT @ I
print(f"点乘全一矩阵：{ans4}")

ans1_flat = ans1.flatten()
ans2_flat = ans2.flatten()
ans3_flat = ans3.flatten()
ans4_flat = ans4.flatten()

print(f"ans1 转为一维：{ans1_flat}")
print(f"ans2 转为一维：{ans2_flat}")
print(f"ans3 转为一维：{ans3_flat}")
print(f"ans4 转为一维：{ans4_flat}")

```
