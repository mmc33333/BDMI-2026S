## I: Calculating Pi

### 用幂级数计算：
$$
\arctan x = x - \frac{x^3}{3} + \frac{x^5}{5} -...+ (-1)^{n-1}\frac{x^{2n-1}}{2n-1}
$$

```python
import numpy as np

def arctan(x, n):
    ans = 0.0
    for i in range(1,n):
        j = 2*i-1
        if i%2 == 1:
            ans += (x**j)/j
        else:
            ans -= (x**j)/j
    return ans

if __name__ == '__main__':
    n_list = [3,5,10,100,1000,100000]
    for n in n_list:
        print(f"pi = {4*arctan(1,n):.5f}, when n = {n}")
```

### 用Monte-Carlo方法：

```python
import numpy.random as npr
import numpy as np

def MonteCarloForPi(n):
    num_in = 0
    for i in range(int(n)):
        x, y = npr.rand(), npr.rand()
        #print(x,y)
        r = np.sqrt(x*x+y*y)
        if r <= 1:
            num_in += 1
    return (4*num_in)/n

if __name__ == '__main__':
    n_list = [10,100,1e7]
    for n in n_list:
        print(MonteCarloForPi(n))
```

## II: Instantiating A Cat

要求：init，color，jump 使我的猫跳两步

```python
class MiCat:
    def __init__(self, color, age):
        self.color = color
        self.age = age
    
    def jump(self):
        print(f"The CAT which is {self.color} and {self.age} years old has just jumped once!!!")

cat = MiCat("black", 0.5)
for i in range(2):
    cat.jump()
```
